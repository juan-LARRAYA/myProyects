```python

#Read the original image
img_read = cv2.imread("test.jpg", cv2.IMREAD_GRAYSCALE)
# Perform global thresholding
retval, img_thresh_gb1_1 = cv2.threshold(img_read,50, 255, cv2.THRESH_BINARY)
#Perform global thresholding
retval, img_thresh_gbl_2 = cv2.threshold(img_read,170, 240, cv2.THRESH_BINARY)
# Perform adaptive thresholding
img_thresh_adp = cv2.adaptiveThreshold(img_read, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 21, 7)
#S Show the images
plt.figure(figsize=[18,15])
plt.subplot(221); plt.imshow(img_read, cmap="gray"); plt.title("Original");
plt.subplot(222); plt.imshow(img_thresh_gb1_1,cmap="gray"); plt.title("Thresholded (global: 50)");
plt.subplot(223); plt.imshow(img_thresh_gbl_2,cmap="gray"); plt.title("Thresholded (global: 130)");
plt.subplot(224); plt.imshow(img_thresh_adp, cmap="gray"); plt.title("Thresholded (adaptive)");

cv2.imwrite("test_with_definition.jpg", img_thresh_gbl_2,[cv2.IMWRITE_JPEG_QUALITY, 100])
cv2.imwrite("test_with_hight_definition.jpg", img_thresh_adp,[cv2.IMWRITE_JPEG_QUALITY, 100])
```
